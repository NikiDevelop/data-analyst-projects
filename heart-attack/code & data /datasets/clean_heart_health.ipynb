# Correlación de los Datos y Visualización

```python
numeric = df.select_dtypes(include='number')
corr_matrix = numeric.corr()
corr_matrix
```

```python
size = ['Glucose(mg/dL)', 'Cholesterol(mg/dL)', 'Weight(kg)', 'Height(cm)', 'Age', 'Exercise(hours/week)']

size_corr = pd.DataFrame(df[size + ['Heart Attack']].corr()['Heart Attack'].sort_values(ascending=False))
size_corr = size_corr[1:]

plt.figure(figsize=(15,4))
sns.set(rc={'figure.facecolor':'white','axes.facecolor':'white'})
plt.title('¿Qué Correlación hay con los Ataques al Corazón?', fontsize=16, fontweight='bold', y =1.1)
plt.suptitle('Correlación que existe entre los atributos',fontsize=14, y=0.95)
sns.set_style("whitegrid")
ax = sns.barplot(x='Heart Attack', y=size_corr.index, palette = "coolwarm_r", data=size_corr)
ax.set_xlabel('Correlación Ataques al Corazón (0 a 1)')



#plt.savefig("images/Size_Corr_Plot.png", dpi=300)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlación de Ataques al Corazón')
plt.show()
```

```python
# plotting grid
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 12))
fig.suptitle('Destribución de las Variables Númericas', fontsize=16)

# density plots
sns.histplot(df['Edad'], kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Distribución por Edad')

sns.histplot(df['Altura(cm)'], kde=True, ax=axes[0, 1], color='salmon')
axes[0, 1].set_title('Distribución por Altura')

sns.histplot(df['Peso(kg)'], kde=True, ax=axes[1, 0], color='lightgreen')
axes[1, 0].set_title('Distribución por Peso')

sns.histplot(df['Colesterol(mg/dL)'], kde=True, ax=axes[1, 1], color='orange')
axes[1, 1].set_title('Distribución por Colesterol')

sns.histplot(df['Glucosa(mg/dL)'], kde=True, ax=axes[2, 0], color='lightcoral')
axes[2, 0].set_title('Distribución por Glucosa')

sns.histplot(df['Ejercicio(horas/semana)'], kde=True, ax=axes[2, 1], color='lightseagreen')
axes[2, 1].set_title('Ejercicio Horas por Semana')


plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))
fig.suptitle('Distribución de las Variables por Categoria', fontsize=16)

sns.countplot(x='Género', data=df, ax=axes[0], palette='Set2')
axes[0].set_title('Género')

sns.countplot(data=df, ax=axes[1], palette='Pastel1')
axes[1].set_title('Fumadores')

sns.countplot(x='Heart Attack', data=df, ax=axes[2], palette='Set3')
axes[2].set_title('Ataques al Corazón')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
